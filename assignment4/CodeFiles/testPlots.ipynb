```python
import torch
import numpy as np
import matplotlib.pyplot as plt
from sklearn.manifold import TSNE
```

```python
testDataPath = '../DataFiles/A4_test_data.pt'
testData = torch.load(testDataPath)
# tests = testData['energy'].reshape(-1, testData['energy'].shape[-1])
tests = testData['energy'].reshape(testData['energy'].shape[0], -1)
tests = tests[::100]
print(tests.shape)

tsne = TSNE(n_components=2, perplexity=30, learning_rate='auto', init='random', random_state=42)
embeddedTest = tsne.fit_transform(tests.cpu().numpy())

plt.figure(figsize=(8, 6))
plt.scatter(embeddedTest[:, 0], embeddedTest[:, 1], s=5, alpha=0.6)
plt.title("t-SNE Projection of Algo1 Samples in 2D")
plt.xlabel("Component 1")
plt.ylabel("Component 2")
plt.grid(True)
plt.savefig('SampledData.png', dpi=300)
plt.show()
```

```python
# I want to read in the sampled data and fit it to the same transform as the about test samples to get a better idea
# of the distribution of the sampled data
sampledDataPath = '../DataFiles/OutputAlgo1.pt'
sampledData = torch.load(sampledDataPath)
# sampled = sampledData['energy'].reshape(-1, sampledData['energy'].shape[-1])
# sampled = sampledData['energy'].reshape(sampledData['energy'].shape[0], -1)
print(sampledData.shape)
```

```python
sampledData = sampledData.reshape(sampledData.shape[0], -1)
embeddedSampled = tsne.fit_transform(sampledData.cpu().numpy())
plt.figure(figsize=(8, 6))
plt.scatter(embeddedSampled[:, 0], embeddedSampled[:, 1], s=5, alpha=0.6)
plt.title("t-SNE Projection of Algo1 Samples in 2D")
plt.xlabel("Component 1")
plt.ylabel("Component 2")
plt.grid(True)
plt.savefig('sampledDataAlgo1.png', dpi=300)
plt.show()
```
